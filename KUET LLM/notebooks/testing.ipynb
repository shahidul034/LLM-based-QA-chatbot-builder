```python
import pandas as pd
f=open("microprocessor lab marks.txt").read().split("\n")
```

```python
dat=[]
sh_10={}
az_10={}
for x in f:
    xx=x.split(" ")
    if len(xx)>=3:
        # az_10[1807000+int(xx[0])]=int(xx[1])
        # sh_10[1807000+int(xx[0])]=int(xx[2])
        
        dat.append({
            "Roll": 2007000+int(xx[0]),
            "marks":int(xx[1])+int(xx[2])
        })
    else:
        print(x)
```

```python
df=pd.DataFrame(dat)
df_new=df.sort_values('Roll')
df_new.to_csv("mc sorted mark.csv",index=False)
```

```python
az_20={}
sh_20={}
az_20_df=pd.read_excel(r"data\CSE 4221(azhar sir).xlsx")
sh_20_df=pd.read_excel(r"data\NLP-2k18 mark(20).xlsx")
name_roll=pd.read_csv(r"data\CSE 4221.csv")
```

```python
import math
roll_email={}
roll_name={}
for x in az_20_df.values:
    az_20[int(x[0])]=int(x[1])
for x in sh_20_df.values:
    m = float(x[3])
    roll_name[int(x[1])]=x[0]
    roll_email[int(x[1])]=x[2]
    if not math.isnan(m):
        sh_20[int(x[1])]=int(x[3])
```

```python
total_marks=[]
for x in name_roll.values:
    roll=int(x[1])
    s10=0
    a10=0
    s20=0
    a20=0
    name="No name"
    email="No email"
    if sh_10.get(roll)!=None:
        s10=sh_10[roll]
    if sh_20.get(roll)!=None:
        s20=sh_20[roll]
    if az_10.get(roll)!=None:
        a10=az_10[roll]
    if az_20.get(roll)!=None:
        a20=az_20[roll]
    if roll_name.get(roll)!=None:
        name=roll_name[roll]
    if roll_email.get(roll)!=None:
        email=roll_email[roll]
    total_marks.append({
        'roll':roll,
        "name":name,
        'email':email,
        "Azhar sir(20)":a20,
        "Azhar sir(10)":a10,
        "Shakib sir(20)":s20,
        "Shakib sir(10)":s10
    })
```

```python
df_fin=pd.DataFrame(total_marks)
df_fin.to_csv("Total_marks_NLP.csv",index=False)
```

```python

f=open("ssbms_schema.sql","r").read()
```

```python
import re

def extract_all_create_tables(query):
    pattern = re.compile(r'CREATE TABLE.*?;', re.DOTALL)
    matches = pattern.findall(query)
    
    return matches

# Example usage:
input_text = f

results = extract_all_create_tables(input_text)
if results:
    print("Found CREATE TABLE statements:")
    for result in results:
        print(result)
else:
    print("No CREATE TABLE statements found.")
```
